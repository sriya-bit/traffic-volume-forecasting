```python
import pandas as pd
import numpy as np

df = pd.read_csv("Integrated_Traffic_Weather_Hyderabad.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
# Removing duplicate rows to avoid biased analysis
```

```python
# Convert DateTime to pandas datetime
df['DateTime'] = pd.to_datetime(df['DateTime'])

# Time-based features
df['DayOfWeek'] = df['DateTime'].dt.dayofweek
df['Month'] = df['DateTime'].dt.month
df['IsWeekend'] = df['DayOfWeek'].isin([5, 6]).astype(int)

df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

numeric_cols = ['Vehicles', 'Temperature', 'Precipitation', 'WindSpeed', 'Humidity']
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

df.head()
```

```python
df.to_csv("Final_Preprocessed_Traffic_Data.csv", index=False)
```
